### 라그랑주 승수법(Lagrange Multiplier)

라그랑주 승수법(미정 승수법 Undetermined Multiplier): 하나 또는 그 이상의 제약 조건하에서 여러 변수에 대한 임계점을 찾는데 사용.

- Q. 함수 $f(x_{1},x_{2})$의 최대값을 찾는 문제

이 때 제약조건을 다음의 형태로 작성. $ g(x_{1},x_{2}) = 0 $ <br>
이를 위한 한 가지 방법은 위 제약 조건식은 풀어 $ x_{2} = h(x_{1}) $의 형태로 표현한 후, $f(x_{1}, h(x_{1}))$의 형태의 x1에 관한 함수를 GET.<br>
그리고 x1의 미분을 활용해 최댓값을 찾는 것이 일반적인 방법이다.

BUT, 이 방법의 문제점은 x2를 x1의 함수로 표현하게 하는는 제약 조건의 Analytic Solution을 찾기 힘들다는 것. 또한 x1과 x2를 다르게 취급하는 방법으로 변수들 간의 자연적인 대칭성(Natural Symmetry)를 망칠 수 있음.

라그랑주 승수인 매개변수 $\lambda$를 활용하면 명쾌하고 보통 단순하게 문제를 해결할 수 있음. 기하학적 관점에서 보면 성분 $x_{1}, ..., x_{D}$를 가지는 D차원의 변수 $\mathbf{x}$에 대해 제약식 $g(\mathbf{x}) = 0$은 $\mathbf{x}$ 공간상의 (D-1)차원 표면에 해당함. 아래의 그림처럼.

제약조건 표면상의 모든 점들에 대해 제약조건함수에 대한 기울기 $\nabla g(\mathbf{x})$들은 표면에 대해 직교할 것을 확인하기 위해 표면상의 점 x와 주변 점 x+e를 고려해보고 x 주변에 대해 테일러 전개를 하면 
$$ g(\mathbf{x} + \epsilon) \simeq g(\mathbf{x}) + \epsilon^{T} \nabla  g(\mathbf{x}) $$
$ \mathbf{x} + \epsilon $ 와 $ \mathbf{x} $ 는 둘다 제약조건 평면상에 놓여 있으므로 $ g(\mathbf{x} + \epsilon) = g(\mathbf{x}) $이며 따라서, $ \epsilon^{T} \nabla  g(\mathbf{x}) \simeq 0 $이다. $\lim \lVert \epsilon \rVert \to 0 $으로 하면, $ \epsilon^{T} \nabla  g(\mathbf{x}) = 0 $으로 e는 제약조건평면 g(x) = 0에 대해 평행하기 때문에 벡터 $ \nabla g$는 표면에 대해 수직임을 알 수 있다.

이제 f(x)가 최대가 되는 제약조건표면상의 점을 찾아보자. 최대가 되게하는 점 역시, $ \nabla f(x)$ 역시 제약조건평면에 직교해야 f(x)값을 조금 움직일 때 증가하지 않기 때문이다. 이에 대한 기하학적 도식은 아래와 같으며 $ \nabla f$와 $\nabla f$는 평행 또는 역평행한 벡터들이어야 하며 다음을 만족하는 매개변수 $\lambda$가 존재할 것이다.
$$ \nabla f + \lambda \nabla g = 0$$
이 때, 0이아닌 람다를 라그랑주 승수라고 하며 이제 라그랑주 함수를 도입하면
$$ L(\mathbf{x},\lambda) \equiv f(\mathbf{x}) + \lambda g(\mathbf{x}) $$
이 되므로 위위식은 $ \nabla_{\mathbf{x}}L = 0 $을 하면 구할 수 있으며 제약조건식 g(x)은 $ \partial L / \partial \lambda = 0 $으로 구할 수 있다. 따라서 제약조건 g(x) = 0 에서 함수 f(x)의 최댓값을 구하기 위해서는 $  L(\mathbf{x},\lambda) $의 임계점을 찾으면 된다. D차원 벡터 x의 경우, 임계점 $\mathbf{x}^{*}$과 람다를 정하는 D+1개의 식을 내놓게 되며 $\mathbf{x}^{*}$에만 관심이 있다면 람다를 굳이 찾을 필요없이 정류공식에서 제거할 수 있어 미정승수법이라고도 하는 것이다.

ex) $f(x_{1},x_{2}) = 1 - x_{1}^{2} - x_{2}^{2} $의 임계점 / 제약조건 $ g(x_{1},x_{2}) = x_{1} + x_{2} - 1 = 0 $

$  L(\mathbf{x},\lambda) =  1 - x_{1}^{2} - x_{2}^{2} + \lambda(x_{1} + x_{2} - 1)  $
이고 다음의 방정식을 얻을 수 있다.<br>
$ -2x_{1} + \lambda = 0 $ <br> $ -2x_{2} + \lambda = 0 $ <br> $ x_{1}+x_{2} = 1 $<br>이를 풀면, 임계점 (x1,x2) = (0.5, 0.5) / 람다는 1을 알 수 있다.

지금까지 g(x)=0의 **등식 제약조건(Equality Constraint)**에서 f(x)를 최대화 문제를 살펴 보았고 이제 $g(\mathbf{x}) \ngeqslant 0 $ 형태의 **부등식 제약조건(Inequality Constraint)** 하에 f(x)를 최대화하는 문제를 풀어보자.<br>
이 때 가능한 종류의 해는 2가지 이다. 첫 번째는 임계점이 $g(\mathbf{x}) > 0 $인 지역에 존재하는 경우인 **비활성(inactive)**과 두 번째는 $g(\mathbf{x}) = 0 $ 경계상에 존재하는 경우인 **활성(active)** 이다. <br>
첫 번째 경우 함수 g(x)는 아무런 역할을 하지 않아 정류조건은 단순히 $ \nabla f(\mathbf{x}) = 0 $이다. 이는 람다는 0인 경우의 식 $ L(\mathbf{x},\lambda) $ 라그랑주 함수의 임계점에 해당한다.<br>
두 번째 경우는 앞에서 살펴본 등식제약조건과 비슷하며, 람다는 0이아닌 식 $ L(\mathbf{x},\lambda) $ 라그랑주 함수의 임계점에 해당한다. <br>
등식제약조건과 다르게 함수 f(x)는 그 기울기가 지역 g(x) > 0 으로부터 지향되는 방향으로 있을 때만 최댓값을 가질 것이기 때문에, 라그랑주 승수의 부호가 중요하다. 어떤 람다가 양수일 때 $ \nabla  f(\mathbf{x}) = -\lambda \nabla  g(\mathbf{x}) $를 갖게 된다. <br>
두 경우 모두 $\lambda g(\mathbf{x}) = 0$이므로 f(x)를 제약조건부등식하에서 최대화하는 문제는 라그랑주함수 $ L(\mathbf{x},\lambda) = f(\mathbf{x}) + \lambda g(\mathbf{x}) $를 다음 조건하에 x와 람다에 대해 최적화함으로써 풀 수 있다.<br>
$ g(\mathbf{x}) \ngeqslant 0  $<br>
$ \lambda \ngeqslant 0  $<br>
$ \lambda g(\mathbf{x}) \ngeqslant 0  $<br>
이 조건을 **KKT조건**이라고 한다.(Karush-Kuhn-Tucker) 만약 우리가 제약조건부등식에서 함수 f(x)에 대해 최소화를 한다고 하면, $ \lambda \ngeqslant 0  $에서 라그랑주함수  $ L(\mathbf{x},\lambda) = f(\mathbf{x}) - \lambda g(\mathbf{x}) $를 x에대해 최소화하는 것을 주의해라.

라그랑주 승수법을 여러개의 등식조건과 부등식 조건으로 확장하는 것은 어렵지 않다. $ g_{j}(\mathbf{x}) = 0 $, $ h_{k}(\mathbf{x}) \ngeqslant 0 $에 대해 f(x)를 최대화는 $ \left\{ \lambda_{j} \right\} $와 $ \left\{ \mu_{k} \right\} $를 도입해 다음 함수를 최적화 하면된다.
$$ L(\mathbf{x}, \left\{ \lambda_{j} \right\}, \left\{ \mu_{k} \right\}) = f(\mathbf{x}) + \sum_{j=1}^{J} \lambda_{j} g_{j}(\mathbf{x}) + \sum_{k=1}^{K} \mu_{k} h_{k}(\mathbf{x})   $$
이 때 $ \mu_{k} \ngeqslant 0 $ 과 $ \mu_{k}h_{k}(\mathbf{x}) = 0 $을 만족해야한다. 제약된 범함수 미분에 대한 확장역시 어렵지 않으며 더 자세한 논의는 Nocedal and Wright(1999)를 참고해라.
