## 부록

### 변분법(Calculas of Variation)

범함수(Functional, $ F\left[ y \right] $) : 함수 $y(x)$를 입력받아 출력값 F를  돌려주는 연산자

범함수의 예시<br>
- 2차원 평면에 그려진 곡선의 길이에서 곡선의 경로 함수
- 연속 변수 $x$에 대한 엔트로피 $H\left[x\right]$ 
- 모든 확률 밀도 함수 $p(x)$에 대해 해당 밀도하에서의 x의 엔트로피 값을 돌려주므로 p(x)의 엔트로피 $H\left[ p \right]$

미분을 활용해 함수 y(x)의 최대 최소 값을 찾는 문제와 비슷하게 변분법에서 $ F\left[ y \right] $를 최대 최소하는 특정함수 $y(x)$를 찾는 문제를 다룸.<br>
- 두 포인트 사이의 가장 짧은 경로는 직선
- 최대 엔트로피 분포는 가우시안 분포

일반 미분법의 법칙을 사용하지 않고 미분 $dy/dx$를 계산한다고 해보면, 변수 x에 대해 작은 변화 $\epsilon$을 만들어 $\epsilon$에 대한 거듭제곱을 전개하여 미분을 계산할 수 있다.<br>
$y(x+\epsilon)= y(x)+ {{dy}\over{dx}}\epsilon + O(\epsilon^{2}) $<br>
그리고 $lim \to 0 $를 취한다. 이와 비슷하게 여러 변수 $y(x_{1}, ... , x_{D})$함수의 편미분 함수는 <br>
$y(x_{1} + \epsilon_{1}, ... ,x_{D} + \epsilon_{D})= y(x_{1}, ... , x_{D}) + \sum_{i=1}^{D} {{\partial y}\over{ \partial x_{i}}}\epsilon_{i} + O(\epsilon^{2}) $<br>
또 비슷하게 범함수의 미분에 대해 정의할 수 있다. 함수 $y(x)$에 대해 작은 변화 $\epsilon \eta(x)$를 가했을 때 범함수 $F\left[ y \right]$가 얼마나 변화하는지 고려하는 것이다. $ \eta(x)$는 x에대한 임의의 함수이며 $F\left[ y \right]$의 $y(x)$에 대한 미분은 ${\delta F}/{\delta y(x)}$ 로 표현할 수 있으며 관계식은
$ F\left[ y(x) + \epsilon \eta(x) \right] = F\left[ y(x) \right] + \epsilon \int {{{\delta F}\over{\delta y(x)}}  \eta(x)  dx} +  O(\epsilon^{2})  $<br>
위 식의 자연스러운 확장이며, F[y]는 연속변수들의 집합(모든 x에 대한 y값)에 종속적이다. 범함수는 함수 y(x)의 작은 변화에 대해서는 정류적이여야하며 이는 
$ \int {{{\delta F}\over{\delta y(x)}} \eta(x)  dx} = 0 $ <br>
이며 임의의 $ \eta(x) $에 만족해야한다. 따라서 범함수의 미분값은 0이어야하며 포인트 $\hat{x}$의 이웃지역을 제외한 다른 모든 곳에서 0을 가지는 섭동 $ \eta(x) $의 경우 범함수의 미분값은 $ x = \hat{x} $에서 0이어야 한다. 하지만 이는 모든 $\hat{x}$ 값의 선택에서 성립 되어야 하므로 범함수의 미분은 모든 x값에서 0이어야 한다.

$ y(x) $의 미분함수 $y'(x), x$에 종속인 함수 $ G(y,y', x) $의 적분으로 정의되는 범함수를 고려해보자. <br>
이를 위와 같이 전개하여 범함수 미분값이 0이라는 성질을 이용하면 ** 오일러 라그랑주(Euler-Langrange) ** 공식을 얻을 수 있다.

$ {{\partial G} \over {\partial y}} - {{d} \over {dx}} {{\partial G} \over {\partial y'}}$<br>
$ G = y(x)^{2} + (y'(x))^{2} $<br>
$ y(x) - {{d^{2} y } \over {dx^{2}}} = 0 $

y(x)이 이차미분은 y(x)의 경계조건을 이용해 푼다.<br>
보통 우리는 G(y,x)의 형태를 띠고 y(x)의 미분에 대해 종속적이지 않은 피적분 함수에 대한 적분으로 범함수를 정의한다. 이경우 정류성이 만족되기 위해 모든 x에 대해 $ \partial G / \partial y(x) = 0 $이기만 하면 된다. <br>
범함수의 확률 분포에 대해 최적화 하려면 확률에 대한 정규화 제약조건을 유지해야하며 이는 **[라그랑주 승수법](https://github.com/NamSahng/Summary)**을 이용해 쉽게 시행할 수 있다. 그 후에 제약조건이 없는 최적화를 시행할 수 있다. 위 결과의 다차원 변수 x의 확장은 어렵지 않으며 더 포괄적인 논의를 알고 싶다면 Sagan(1969)를 보아라.


c.f. d Δ δ ∂ 이 넷의 공학 내에서의 차이

1. d: 변화경로에 무관한(path independent)한 property의 differential change양을 나타낼 때 사용하는 거고,
2. 대문자 델타: 이것도 변화량이긴 한데, 미분 적용하는 중간과정에 나오는 것 혹은 라플라스 방정식의 라플라시안 연산자로 보는 경우도 있고,
3. 소문자 델타: 이 경우는 관찰하고자 하는 양이 path dependent 한 것, 이 경우 관찰량은 property가 아니라 열이나 일 같은 경우가 해당되지. 그러나, 중력에 대한 일 같은 경우(비 보존력이 무시되는 경우 혹은 등엔트로피 과정의 경우)에는 이 델타가 d로 변할 수 있지.
4. 라운드 d는 편미분 연산자인데, 어떤 상태량이 여러 변수에 동시에 영향을 받는 경우, 한 변수에 의한 영향만을 볼 때 이 연산자를 쓰지.
