####  Gaussian Elimination

Forward Elimination $\to$ Triangular System (Upper Triangular Matrix) $\to$ Back-substitution

Do pivoting when Zero appears in a pivot position. When all pivots are non-zero after G.E then It has unique solution.

#### Elementary Matrix in Elimination Steps

- Elemantary Matrix<br>
$ E_{21} = \begin{bmatrix}
1 & 0 & 0 \\
-l_{21} & 1 & 0 \\
0 & 0 & 1 \\
\end{bmatrix}$ &nbsp; &nbsp;  $ E_{21}^{-1} = \begin{bmatrix}
1 & 0 & 0 \\
l_{21} & 1 & 0 \\
0 & 0 & 1 \\
\end{bmatrix}$

#### Triangular Factors and Row Exchanges

For each Elementary Matrix
$ E_{32}E_{31}E_{21} A = U $<br>
$ A =  E_{32}E_{31}E_{21} =  E_{21}^{-1}E_{31}^{-1}E_{32}^{-1}  U = LU $<br>

$\to$ Triangular Factorization, LU Decomposition, LU Factorization

L has history of G.E. det(A) = det(L) * det(U)
}

 LU Factorization are uniquely determined by A. (LU 분할이 되면, 시스템은 일대일 대응이다!)
 
- LDU Factorization

$ U = {\begin{bmatrix}
d_{1} &  &  \\
 & d_{2} &  \\
 &  &  \ddots \\
 &  &  & d_{n} 
\end{bmatrix}}
{\begin{bmatrix}
1 & u_{12} / d_{1} & u_{13} / d_{1} & \cdots  \\
 & 1 & u_{23} / d_{2} & \cdots \\
 &  &  \ddots & \cdots \\

$

- Row Exchanges (Permutation Matrix): P has the same rows as the I, Single '1' in every row and column.

$ P_{21} = \begin{bmatrix}
0 & 1 & 0 \\
1 & 0 & 0 \\
0 & 0 & 1 \\
\end{bmatrix}  $


$ P^{-1 } = P $, 교환법칙 성립 안함

#### Inverses and Transposes

Note 1. The inverse exists if elimination produces n pivots. (Row exchanges are allowed)<br>
Note 2. The Inverse matrix is unique <br>
Note 3. If A is invertible, 1:1 대응이다. <br>
Note 4. if Det(a) is not 0, then A is invertible


- Gausss-Jordan Method<br>
$\left[A \ E \right]$ $\to$ $\left[U \ L^{-1} \right]$ $\to$ $\left[I \ \ \ U^{-1} L^{-1} \right]$

- Symmetric Metrix (대칭행렬)<br>
$ A^{T} = A $<br>
If A is Symmetric then it is Invertible<br>
If A is Symmetric then $ L^{T} = U $

ex) Correlation Matrix: 상대방 성분을 내가 얼마나 갖고 있는지(상관성), 내적과 같다. 이를 확장해 Eigen Vector, Eigen Value <br>
