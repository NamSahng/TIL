### Chapter 4 Determinants


#### Important properties of det(A)  

1) $ A^{-1} $ exissts if $ det(A) \neq 0 $ <br>
2) det A equals volume of a box in $\mathbf{R}^{n}$ (공간의 부피를 의미한다.) <br>
  ex) xyz coordinate $ \to $ cylindrical coordinate로 변환하여 적분 할 때 ($ x = rcos\theta, y = rsin\theta, z = z  $) , <br>
  $ dxdydz  \to drd\theta dh $에서  $  dxdydz  = J drd\theta dh  $ J 라는 야코비안 행렬의 determinant를 곱하고 적분한다. <br>
  Jacobian $ J = \begin{vmatrix}   \partial x / \partial r  & \partial x / \partial \theta  & \partial x / \partial z \\   \partial y / \partial r  & \partial y / \partial \theta  & \partial y / \partial z \\  \partial z / \partial r  & \partial z / \partial \theta  & \partial z / \partial z \\  \end{vmatrix} =     \begin{vmatrix}   cos\theta  & -rsin\theta  & 0 \\  sin\theta  & rcos\theta  & 0 \\    0  & 0  & 1 \\  \end{vmatrix} = r $ <br>
  이와 유사하게 spherical coordinate도 생각해 볼 수 있다.<br>
3) det A = G.E의 Upper Triangular Matrix의 product of pivots (row exchange에 따라 부호 바뀜.) <br>
4) Cramer's Rule <br>
Ax = b 의 해 $ x_{i} = {{det( A_{i}  )}\over {det(A)}} $ <br>
$ A_{i} = \begin{bmatrix} a_{1} &  a_{2} & \cdots & a_{i-1} & b & a_{i+1} &   \cdots & a_{n} \end{bmatrix}  $

#### Properties of Determinant

- Basic 3 <br>
1) The determinant of identity matrix is 1. <br>
2) The determinant of sign changes whe 2rows are exchanged. <br>
3) The determinant depends linearly on the first row. <br>
ex) $ \begin{vmatrix} a + \grave{a} & b + \grave{b} \\ c & d \end{vmatrix} =  \begin{vmatrix} a  & b  \\ c & d \end{vmatrix} + \begin{vmatrix} \grave{a} & \grave{b} \\ c & d \end{vmatrix} $ &nbsp; &nbsp; &nbsp;
$ \begin{vmatrix} t a  &  tb  \\ c & d \end{vmatrix} = t \begin{vmatrix} a  & b  \\ c & d \end{vmatrix} $ <br>


- Additional Properties by the basics. <br>
4) If 2 rows are equal, then det A = 0  <br>
5) Subtracting a multiple of one row from another row(row operation) leaves the same determinant. <br>
$ \begin{vmatrix} a -lc & b -ld \\ c & d \end{vmatrix} = \begin{vmatrix} a & b \\ c & d \end{vmatrix} + \begin{vmatrix} -lc & -ld  \\ c & d \end{vmatrix}  $  <br>
6) If A has a row of zeros, then det A = 0 (column) <br>
7) If A is triangular, det A is product of diagonal entries. <br>
8) If A is singular then det A = 0 , If A is not-singular(invertible) then det A is not 0 (but if it is close to 0, it is called ill-conditioned(문제가 있음))   <br>
9) $ det(AB) = det(A) \cdot det(B) $ <br>
10) $ det(A) = det(A^{T} )$


#### Formulas for the Determinants

- Big formula: how to get determinant <br>
$$ det A = \sum_{all P's} (a_{1\alpha}a_{2\beta} \cdots a_{n\nu}) det(P) $$ 
$ det (P) = +1 \ or \ -1 $ for an even or odd number of row exchnages.

- Expansion of det A in Cofactors: focous on one row <br>
$ det(A) = a_{i1}C_{i1} + a_{i2}C_{i2} + \cdots + a_{in}C_{in} $ <br>
$ C_{ij} = (-1)^{i+j}det (M_{ij}) \quad $ M is minor matrix, sub matrix that deleted ro i and column j <br>
<img  src="./image/image_4.2.PNG" width="70%">
ex) <br>
$ A =  \begin{bmatrix} 
2 & -1 & 0 & 0 \\ 
-1 & 2 & -1 & 0 \\
0 & -1 & 2 & -1 \\
0 & 0 & -1 & 2 \end{bmatrix} $ <br>
$ det A = a_{11}(-1)^{1+1} det ( M_{11} ) + a_{12}(-1)^{1+2} det ( M_{12} ) $ <br>
$ \quad \quad = 2 \cdot \begin{vmatrix} 2 & -1 & 0 \\ -1 & 2 & -1 \\  0 & -1 & 2 \end{vmatrix} + 1 \cdot \begin{vmatrix} -1 & -1 & 0 \\ 0 & 2 & -1 \\  0 & -1 & 2 \end{vmatrix} = 5 $ <br>
실제로는 G.E.를 사용하는 것이 더 빠름.

#### Applications of Determinants

1) Computation of $A^{-1}$ <br>
cofactor Matrix C를 통해 구할 수 있다. <br>
cofactor Matrix $ \mathbf{C} = \begin{bmatrix} C_{11} & \cdots & C_{1n} \\ \vdots & \ddots & \vdots \\ C_{n1} & \cdots & C_{nn} \end{bmatrix} $ <br>
$ A^{-1} = {{\mathbf{C}^{T}}\over{det(A)}} \quad A^{-1}_{ij} = {{\mathbf{C}_{ji}}\over{det(A)}} $ <br>
하지만, C matrix를 구하는게 빡세서 GE로 구하는게 더 빠름.
