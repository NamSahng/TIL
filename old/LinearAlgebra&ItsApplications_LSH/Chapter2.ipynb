### Chapter 2 Vector Spaces

Mainly about when \# of Equation < \# of Unknowns

Space: x and y should be closed under addition and scalar multiplication<br>
$ x, y \in \mathbf{V} , c_{1}x + c_{2}y \in \mathbf{V} $<br>
There is a unique "Zero Vector" such that x + 0 = x  . $\forall$ x (원점을 포함하지 않으면 벡터 공간이 될 수 없다.)<br>

Subspace: non empty subset that satisfies the requirements for a vector space.<br>
The origin is smallest subspace.

- The Column space of A<br>
Column space: Set of all Linear combination of columns of Matrix A. $\mathbf{C}(A)$ <br>
$ Ax = b \ \to \ \begin{bmatrix}  | & | &  & | \\ a_{1} &  a_{2} & \cdots & a_{n} \\
 | & | &  & | \end{bmatrix} \begin{bmatrix} x_{1} \\ x_{2} \\ \vdots \\ x_{n} \end{bmatrix} = \begin{bmatrix} b_{1} \\ b_{2} \\ \vdots \\ b_{n} \end{bmatrix} $ <br>
$ \to x_{1} \left[ a_{1} \right] + x_{2} \left[ a_{2} \right] + ... + x_{n} \left[ an \right] = \left[ b \right] $ <br>
Ax=b is solvable if b can be expressed as a combination of the columns of A. (b i in column space.) Those Vector b in the spanned space of column vectors of A?

- Nullspace of A<br>
Nullspace of Matrix A consists of all vectors x such that Ax=0 $\mathbf{N}(A)$

#### Solving Ax=0 and Ax=b

Any vector $x_{n}$ in the nullspace can be added to a particular solution $x_{p}$. The solutions to all linear equations have this form, $ x = x_{p} + x_{n} $<br>
Complete solution $ Ax_{p} = b $ and $ Ax_{n} = 0 $ produce  $ A(x_{p}+x_{n}) = b $ 

When the column space doesn't contain every b in $R^{m}$, we need the conditions on b that make $Ax = b$ solvable.

 - Pivot Variables and Free Variables <br>
For solving Rx = 0, Pivots are important <br>
$ Rx = \begin{bmatrix}  1 & 3 & 3 & 2 \\  0 & 0 & 1 & 1 \\  0 & 0 & 0 & 0 \end{bmatrix} \begin{bmatrix} u \\ v \\ w \\ y \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \\ 0 \end{bmatrix} $ <br>
Pivot variables with pivot columns : u, w / Free variables without pivot columns : v, y<br>

#### Linear Independence, Basis, and Dimension

- Linear Independence<br>
식 $ c_{1}v_{1} + c_{2}v_{2} + \cdots + c_{n}v_{n} = 0 $ 에서 모든 $ c_{i} = 0 $ 일 때만 성립하면 벡터들 끼리 서로 Linearly independent하다고 한다. 벡터들 끼리 dependent 하다는 것은 One vector is  a combination of the others 라는 것이다.<br>
If Gaussian Elimination of A generates m non-zero rows then there are m of independent column vectors in A

- Spanning <br>
 All linear combination of vectors $ \left\{ v_{1},v_{2}, ..., v_{n} \right\} $
 
- Basis (Vectors)<br>
 number of minimum linearly independent vectors to span the vector space<br>
 이 때 linear combination is unique from basis (일대일 대응) 하지만 vector space에서 basis는 unique 하진 않다.<br>
 또한 basis vector 들 끼리 Orthogoanl하면 좋다.(Chapter 3)
 maximal independent set & minimal spanning set.

- Dimension of a Vector Space<br>
 number of basis vectors = number of linearly independent vectors<br>
 infinitely many different bases.<br>
 degree of freedom <br>

- Rank of A <br>
 = # of independent column vectors <br>
 = # of independent row vectors <br>
 = # of pivots in Gaussian Elimination of A (independent column vectors 를 찾는 방법) <br>
 = Dim of $ \mathbf{C}(A)2$, ..., v_{n}
 


#### 4 Fundamental Subspaces

A: n x m Matrix에서 <br>
1) The Column space, $ \mathbf{C}(A) $ : Linear Combinations of column vectors, Its dimension is the rank r <br>
2) The Null space, $ \mathbf{N}(A) $ : $ \left\{ x | Ax = 0 \right\} $ Its dimension is the n-r <br>
3) The Row space, $ \mathbf{C}(A^{T}) $ : Linear Combinations of row vectors, Its dimension is also r <br>
4) The Left Null space, $ \mathbf{N}(A^{T}) $ :  $ \left\{ y | A^{T}y = 0 \right\} $ Its dimension is rank m-r <br>
 
$ \mathbf{N}(A) $ and  $ \mathbf{C}(A^{T}) $  are subsets of $  \mathbf{R}^{n} $<br>
$ \mathbf{N}(A^{T}) $ and  $ \mathbf{C}(A) $  are subsets of $  \mathbf{R}^{m} $ (다 포함하는 것은 아니다.)

$ \mathbf{C}(A) \bot \mathbf{N}(A^{T}) $ and $ \mathbf{C}(A)^{T} \bot \mathbf{N}(A) $

- Existenc of Inverses <br>
A: m x n Matrix 에서 <br>
$ A A^{-1} = I$(mxm matrix)  이 되는 $ A^{-1} $을 Right Inverse (m < n) <br>
$ A^{-1} A  = I$(nxn matrix)  이 되는 $ A^{-1} $을 Left Inverse (m > n)  라고 하는데 <br>
Ax = b 문제에서 left inverse가 존재한다 하면 ( A에서 미지수 < 식) $ A^{-1}Ax = A^{-1}b $로 unique한 해를 가짐을 알 수 있으며 이는 기하학적으로 선들이 한점에서 만남을 의미. (원래는 해를 갖기 힘든 부분. 그래서 나중에 Least Square 로 ) <br>
Ax = b 문제에서 right inverse가 존재한다 하면 ( A에서 미지수 > 식) $ AA^{-1}x = A^{-1}b 을 만족하는데 $ A^{-1}b $를 만족하는 벡터들은 많이 만들 수 있어 따지기 어렵다. 그래서 무수히 많은 해를 가짐
