## Outlier
<hr></hr>

- Novelty
    - 데이터의 본질적 특성은 같지만 유형이 다른 관측치
    - 약간의 긍정적으로 해석 가능


- Anomaly
    - 대부분의 데이터와 특성이 다른 관측치
    - 약간의 부정적으로 부정적으로 해석 가능
    

- Outlier
    - 대부분의 데이터와 본질적인 특성이 다른 관측치
    - 굉장히 부정적인 의미
    - 실제로 학습에 부정적인 영향까지 미칠 수 있는 관측치로 Novelty나 Anomaly의 제거는 선택일 수 있으나 Outlier는 제거를 목적으로 찾는 관측치
    


## LOF (Local Outlier Factor)
<hr></hr>

- Outlier Score(Degree of Outlier) 계산 가능.
    - 관측치의 score가 어떤 threshold를 넘었을 때 이상치인지에 대한 언급은 없음.


- 밀도가 짙은 부분은 Outlier Score(LOF)가 작게 나오고, 옅은 부분은 크게 나옴.
- <img style="float:left; margin-right:100%; margin-bottom:1.5%;margin-top:1.5%;" width="400" src="https://upload.wikimedia.org/wikipedia/commons/5/59/LOF.svg">


&nbsp;
### - K-distance of object p 
<hr></hr>

- p 관측치에 대하여, 자신을 제외하고 K번째로 가까운 이웃과의 거리
- <img style="float:left; margin-right:100%; margin-bottom:1.5%;margin-top:1.5%;" width="200" src="./img/lof_1.png">

&nbsp;


### - K-distance neighbor of object p: $N_{k}(p)$
<hr></hr>

- k번째 가까운 이웃과의 거리를 원을 표현할 때, 원 안에(원 위에) 포함되는 관측치의 수
- <img style="float:left; margin-right:100%; margin-bottom:1.5%;margin-top:1.5%;" width="200" src="./img/lof_2.png">

&nbsp;


### - Reachability Distance: $\text{Reachability-distance}_{k}(p,o)$
<hr></hr>

- o를 기준으로 k번째 가까운 이웃과의 거리와 o와 p사이 거리 간의 최대값
- $Reachability-distance_{k}(p,o) = \max{\{\text{K-distance of o}, d(p,o)\}}$
- 정상 관측치인 경우, $\text{K-distance of o}, d(p,o)$ 모두 짧아 Reachability Distance가 짧을 것이라고 볼 수 있다.
- 이상치인 경우 $d(p,o)$ 가 커져 max값 인 Reachability Distance는 클 수 있다. 
- ex. k=5일 때 
    - $\text{Reachability-distance}_{5}(p,o) = \max{\{1.2, 2 \}} = 2$
    - <img style="float:left; margin-right:100%; margin-bottom:1.5%;margin-top:1.5%;" width="200" src="./img/lof_3.png">


&nbsp;

### - Local Reachability density of object p: $\text{lrd}_{k}(p)$
<hr></hr>

- 여러 reachability distance를 하나의 지표로 계산한 값
- $lrd_{k}(p) = {{|N_{k}(p)|}\over{\Sigma_{o\in N_{k}(p)} \text{reachability distance}_{k}(p,o)}}$
    - o 는 K-distance neighbor of object p에 포함되는 data들
- p의 주변의 밀도가 높은 경우 분모가 작아져 $\text{lrd}_{k}(p)$ 값이 커지게 됨 $\to$ 정상

&nbsp;

### - Local Outlier Factor of object p: $LOF_{k}(p)$
<hr></hr>

- 어떤 점 p의 최종 LOF 값 계산
- $LOF_{k}(p) = {{\Sigma_{o\in N_{k}(p)} \ lrd_{k}(o) / \ lrd_{k}(p) }\over{|N_{k}(p)|}}$
    - o 는 K-distance neighbor of object p에 포함되는 data들
- p의 주변 밀도가 높은 경우: LOF 값은 1과 가까워짐
    - $\text{lrd}_{k}(p)$가 크고 $\text{lrd}_{k}(o)$가 큼
- p의 주변 밀도가 낮은 경우: LOF 값은 1보다 많이 커짐
    - $\text{lrd}_{k}(p)$가 작고 $\text{lrd}_{k}(o)$가 커짐
- p의 주변 밀도는 낮지만 정상 군집 내에 있는 것 처럼 보이는 경우: LOF 값은 1과 가까워짐
    - $\text{lrd}_{k}(p)$가 작고 $\text{lrd}_{k}(o)$도 작아짐
      
&nbsp;

### - Issues
<hr></hr>

- K값을 어떤 값으로 정해야하는지
- LOF값을 어떤 임계치에서 이상치로 정의할 것인지
- 고차원에서의 거리
- scaling/normalizing을 해야하는지


## Isolation Forest
<hr></hr>

- Isolation Forest(IF)는 모델 기반으로 보거나 밀도 기반의 알고리즘으로도 볼 수 있다.
- 의사결정나무(DT) 모델과 거의 같지만, 완전히 일치하지는 않음

- DT와 IF 모두 리프노드(터미널 노드)에서 결정이 이루어지기 때문에 중요하다.

- Isolation Foreset
    - 한 객체를 고립시키는 의사결정나무 모델을 구축하는 방법론
        - 정상 관측치를 고립되게 하기 위해서는 여러 번 분할해야 함.
        - 이상 관측치를 고립되게 하기 위해서는 적게 분할해도 됨.
        - <img style="float:left; margin-right:100%; margin-bottom:1.5%;margin-top:1.5%;" width="500" src="./img/IF_1.png">    
    - DT에서는 MSE나 엔트로피 기반으로 분할 변수와 분할점을 찾았다.
        - IF에서는 랜덤하게 트리를 여러번 분할한다.
        - 각각의 트리에서 몇 번만에 고립되었는지 분리 횟수를 계산해 평균을 낼 수 있다.
    - (고립시키기 위한) 분리 횟수(path length)를 기반으로 이상치 점수 계산
    - 이상치 점수: $s(x,n) = 2^{-{{E(h(x))}\over{c(n)}}}$
        - $E(h(x)$: 각 트리들의 average path length, $h(x)$: paht length
        - $c(n)$: 모든 관측치의 평균 path length의 추정치
        - 1에 가까울 수록 이상 데이터, 0에 가까울수록 정상 데이터
        - 어떤 데이터가 평균 path length와 가깝다면, 0.5와 비슷하다.




##  PCA, AE, GAN
<hr></hr>

- 재구축 오차(Reconstruction Error) 기반 방안
    - 간단하고, 성능이 좋음
    - 재구축 오차: 원본데이터와 압축된 데이터를 복원한 데이터의 차이
    - 아래와 같이 정상데이터는 복원이 잘되고, 비정상데이터는 복원이 잘 되지 않는다는 가정을 기반으로 함.
    - 압축 알고리즘은 정상데이터만으로 구축해야 잘 작동할 것으로 예상할 수 있다.

- 여러 재구축 기반 이상치 탐지 알고리즘

    - AE
    - LSTM-AE
    - EncDec-AD
    - VAE
    - Adversarial AE

        - Input을 저차원으로 요약된 행렬(Low ranke Matrix)과 이상치(Sparse Matrix)로 분리

- GAN 기반

    - Generator는 정상데이터로만 학습했기 때문에, 이상데이터의 입력이 들어오면 큰 차이를 낼 것이라는 가정.
